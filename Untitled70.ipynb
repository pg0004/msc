```python
############################
bad_N =['6','8','1''3','2','5','.','-']
df['special_shipping_offer']=df['special_shipping_offer'].astype(str).str[0].replace('$','').replace('9','').replace('`',''). replace(
    bad_N ,'').replace('3','').replace('1','').replace('','NAN').replace('7','')
                                                                    
df['special_shipping_offer']
```

```python
df['special_shipping_offer']= df['special_shipping_offer'].replace('P','pick up').replace('U','Unknow').replace(
    'f','free shipping').replace('s','download').replace('p','pick up').replace('d','digit access').replace(
 'o','digit access').replace('S','pick up').replace('v','variable shipping').replace('C','pick up').replace(
    'A','pick up').replace('m','free shipping').replace('e','pick up').replace('i','pick up').replace(
    'w','free shipping').replace('O','pick up').replace('c','free shipping').replace('a','pick up').replace(
    'N','pick up').replace('F','free shipping').replace('l','digit access')
                                                                                                               
```

```python
value_counts=df['special_shipping_offer'].value_counts()
```

```python
to_remove = value_counts[value_counts < 20].index
df['special_shipping_offer']=df['special_shipping_offer'][~df.special_shipping_offer.isin(to_remove)]
```

```python
# remove bad character function
def remove(list): 
    list = [''.join(x for x in i if x.isalpha()) for i in list]
                    
    return list
```

```python
ss =remove(df['why_buy'])
```

```python
# extract the first few letters as signal represent of
# why_buy variable
sig=[]
for z in ss:
    sig.append(z[0])
print(sig)
```

```python
df['why_buy']=sig
df['why_buy'].unique()
```
