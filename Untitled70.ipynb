```python
def ordinal_encoding(df,col,mapping):
    ordinal_encoder = ce.OrdinalEncoder(cols = [col],return_df = True,mapping = [{'col':col,'mapping':mapping}])
    df_final = ordinal_encoder.fit_transform(df)
    return df_final
```

```python
df = ordinal_encoding(df,'promoted',{'Not Promoted':0,'Promoted':1})
df.promoted.value_counts()
```

```python
target = df['promoted']
cat = df[['level1_category_name','level2_category_name','level3_category_name','holiday_display','why_buy',
          'special_shipping_offer','season_display']]
cols= cat.columns
```

```python
x_train, x_test, y_train, y_test = train_test_split(cat,target,test_size=0.25)
```

```python
columns = [i for i in x_train.columns]
```

```python
## using k-fold select the train set mean value
kf = KFold(random_state=1, n_splits =10, shuffle= True)
```

```python
x_train_te = x_train.copy()
x_train_te['target'] =y_train
```

```python
all_set = []

for train_index, val_index in kf.split(x_train_te):
    # Create splits:
    train, val = x_train_te.iloc[train_index], x_train_te.iloc[val_index]
    val=val.copy()
    
    # Calculate the mean of each column:
    means_list = []
    for col in columns:
        means_list.append(train.groupby(str(col)).target.mean())
    
    # Calculate the mean of each category in each column:
    col_means = []
    for means_series in means_list:
        col_means.append(means_series.mean())
    
    # Encode the data:
    for column, means_series, means in zip(columns, means_list, col_means):
        val[str(column) +'_target_enc'] = val[str(column)].map(means_series).fillna(means) 
    
    list_of_mean_enc = [str(column) + '_target_enc' for column in columns]
    list_of_mean_enc.extend(columns)
    
    all_set.append(val[list_of_mean_enc].copy())

x_train_te=pd.concat(all_set, axis=0)
```

```python
x_train_te
```

```python
x_val_te = pd.DataFrame(index=x_test.index)
for column, means in zip(columns, col_means):
    enc_dict = x_train_te.groupby(column).mean().to_dict()[str(column) + '_target_enc']
    x_val_te[column] = x_test[column].map(enc_dict).fillna(means)
```
